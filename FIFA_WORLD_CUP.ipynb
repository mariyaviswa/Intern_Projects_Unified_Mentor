```python
# Visualizing the Total wins each country
df.Winner.value_counts().sort_values().plot(kind='barh', figsize=(10,5), title="Each Country's Winning counts", alpha=0.7, color='g', xlabel='Total Wins', ylabel='Countries')
```

```python
# Total number of goals scored in world matches from 1930 to 2014
print(sum(df['GoalsScored']), "goals scored in world matches from 1930 to 2014")
```

```python
plt.bar(df.Year, df.GoalsScored)
plt.title('Goals Scored per Year')
plt.xlabel('Year')
plt.ylabel('Goals Scored')
plt.show()
```
