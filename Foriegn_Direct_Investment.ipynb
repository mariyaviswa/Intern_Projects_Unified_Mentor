```python
# Create Average Exchange list
Rates = [45.68, 47.69, 48.39, 45.95, 44.93, 44.27, 45.24, 40.26, 45.99, 47.44, 45.56, 47.92, 54.40, 60.50, 61.14, 65.46, 67.07]
```

```python
# Convert FDI's value USD to INR
def change_value(data, col, num):
  for i in col:
    # Check if the column exists before accessing it
    if i in col:
      data[i] = data[i] * Rates[col.index(i)]/10
    else:
      print(f"Warning: Column '{i}' not found in DataFrame.")
  return data
```

```python
FDI_USD = data.copy()
data1 = change_value(data, years, Rates)
```

```python
data1.style.set_caption("FDI INFLOWS - AMOUNT IN CRORES").format(precision=2)
```

```python
# short the data frame from wide to long format
short = pd.melt(FDI_USD, id_vars=Sectors, value_vars=years, var_name='Year',
                value_name='FDI - (IN US($) MILLION)', ignore_index=True)
short.head()
```

```python
# short the data1 data frame from wide to long format
short1 = pd.melt(data1, id_vars=Sectors, value_vars=years, var_name='Year',
                value_name='FDI - (IN Crores)', ignore_index=True)
short1 = round(short1, 2)
short1.head()
```

```python
# merge the short and short1
Merged = short1.merge(short, how='left')
Merged
```

```python
# Sort the sector and years
sorted_sectors = Merged.sort_values(['Sector', 'Year'], ignore_index=True)
sorted_sectors
```

```python
# Sector wise Total FDI
sorted_FDI  = sorted_sectors.groupby('Sector').sum()
sorted_FDI.sort_values(by = "FDI - (IN US($) MILLION)", ascending=False)
sorted_FDI.drop('Year', axis = 1, inplace = True)
sorted_FDI
```

```python
sorted_FDI.plot(kind='bar', y='FDI - (IN Crores)', figsize = (20, 7), legend=True, title = 'Sector wise FDI', ylabel='FDI in crores')
plt.show()
```

```python
# Top 10 sectors value wise
Top_10_sectors = sorted_FDI.sort_values(by = "FDI - (IN Crores)", ascending=False).head(10)
Top_10_sectors
```

```python
# Top 10 sectors percentage wise
Top_10 = sorted_FDI.nlargest(10, ["FDI - (IN Crores)"])
Total_FDI = round(short1['FDI - (IN Crores)'].sum(), 2)
sum = Top_10['FDI - (IN Crores)'].sum()
Top_10['Percentage'] = round(Top_10['FDI - (IN Crores)'] / sum * 100, 2)
Top_10['% to Total'] = round((Top_10['FDI - (IN Crores)'] / Total_FDI)*100, 2)
Top_10
```

```python
# show the top 10 using bar chart
Top_10.plot(kind='bar', y='FDI - (IN Crores)', figsize = (20, 6), legend=True, title = 'Top 10 Sectors', ylabel='FDI in crores')
plt.show()
```

```python
# show the top 10 using pie chart
plt.figure(figsize=(20,10))
plt.pie(Top_10['FDI - (IN Crores)'], labels=Top_10.index, autopct='%1.2f%%', shadow=False, startangle=90)
plt.axis('equal')
plt.title('Top 10 Sectors')
plt.show()
```

```python
# Least 10 sectors
Last_10_sectors = sorted_FDI.nsmallest(10, ["FDI - (IN Crores)"])
Total_FDI = round(short1['FDI - (IN Crores)'].sum(), 2)
sum = Last_10_sectors['FDI - (IN Crores)'].sum()
Last_10_sectors['Percentage'] = round(Top_10['FDI - (IN Crores)'] / sum * 100, 2)
Last_10_sectors['% to Total'] = round((Last_10_sectors['FDI - (IN Crores)'] / Total_FDI)*100, 2)
Last_10_sectors
```

```python
# show the Last 10 using bar chart
Last_10_sectors.plot(kind='bar', y='FDI - (IN Crores)', figsize = (20, 6), legend=True, title = 'Last 10 Sectors', ylabel='FDI in crores')
plt.show()
```

```python
# show the Last 10 using pie chart
plt.figure(figsize=(20,8))
plt.pie(Last_10_sectors['FDI - (IN Crores)'], labels=Last_10_sectors.index, autopct='%1.1f%%', shadow=False, startangle=90)
plt.axis('equal')
plt.title('Last 10 Sectors')
plt.show()
```
